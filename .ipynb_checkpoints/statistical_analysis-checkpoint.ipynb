## <font color=#FF5733>Importing the necessary Libraries<font/>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import boto3
from io import BytesIO
import scipy.stats as stats
```

## <font color=#FF5733>Accessing the data<font/>

```python
s3 = boto3.client("s3")

bucket_name = 's3bucketaimt'
object_key = "stockdata (1).csv"
```

```python
file_obj = s3.get_object(Bucket=bucket_name, Key = object_key)
df = pd.read_csv(file_obj["Body"])
df.head()
```

## <font color=#FF5733>Data Exploration<font/>

```python
df.head()
```

### <font color="Green">Checking the data types<font/>

```python
df.info()
```

<font color=#0000FF>We notice that the Date column is stored as object. Further down the analysis process, the format type may hinder date-wise analysis or visualization. Hence we will attempt to transform the date variable into a datetime format<font/>

### <font color="Green">Data Transformation<font/>

```python
df['datetime'] = pd.to_datetime(df['Date']) # using datetime format to transform the date variable
                                            # from object to datetime format
```

```python
df.info
```

<font color="blue">The above output shows that the team successfully created a new variable called "datetime" which stores the date in datetime format. It is also a relief to know that there are no missing values in the dataset. So the team will not have to channel any efforts towards imputation<font/>

```python
df["datetime"].nunique()
```

```python
df["month"] = df["datetime"].dt.strftime('%B')
```

```python
df.head()
```

```python
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
```

```python
df["month"] = pd.Categorical(df["month"], categories=month_order, ordered=True)
```

```python
df.info()
```

```python
grouped = df.groupby(["month", "Company"])
aggregated = grouped["Volume"].sum()
```

```python
aggregated
```

```python
pivot_table = aggregated.unstack(level="Company")
pivot_table
```

```python
pivot_table.plot(kind="bar", figsize=(25, 12))
plt.xlabel("Month", fontsize=25)
plt.ylabel("Volume (Millions)", fontsize=25)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.legend(fontsize=18)
plt.show()
```

### <font color="Green">Inferential Statistics<font/>

```python
## IBM values
ibm = df[df["Company"]=="IBM"]

## Microsoft values
msft = df[df["Company"]=="MSFT"]
```

```python
ibm_mean = ibm.mean(numeric_only=True).round(2)
ibm_mean
```

```python
msft_mean = msft.mean(numeric_only=True).round(2)
msft_mean
```

```python
t_statistic, p_value = stats.ttest_ind(ibm["High"], msft["High"])
print("Independent Samples T-Test")
print(f"t-statistic: {t_statistic}")
print(f"p-value: {p_value}")
```
