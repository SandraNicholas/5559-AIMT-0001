```python
import yfinance as yf
import pandas as pd

class StockDataExtractor:
    def __init__(self, ticker):
        self.ticker = ticker
        self.data = None
    
    def fetch_data(self, start_date, end_date):
        self.data = yf.download(self.ticker, start=start_date, end=end_date)
    
    def display_data(self):
        if self.data is None:
            print("Data not fetched yet. Please call fetch_data() method first.")
        else:
            print(self.data)

# Create an instance of the StockDataExtractor class
stock1 = StockDataExtractor("IBM")
stock2 = StockDataExtractor("MSFT")

# Fetch data for the specified time period
start_date = "2021-01-01"
end_date = "2021-12-31"
stock1.fetch_data(start_date, end_date)
stock2.fetch_data(start_date, end_date)
df1 = stock1.data
df1=df1.assign(Company='IBM')
df2 = stock2.data
df2=df2.assign(Company='MSFT')

# Combine the dataframes vertically
combined_df = pd.concat([df1, df2], axis=0)

# Save the combined data to a CSV file
combined_df.to_csv("data.csv")
```
